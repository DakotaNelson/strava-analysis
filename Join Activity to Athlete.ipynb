# Join Activity to Athlete

This notebook shows how to look up an athlete given an activity.

```python
import pymongo
client = pymongo.MongoClient()
db = client.strava
```

```python
join_pipeline = [
    {"$sample": {"size": 1}}, # pick a random activity
    {"$lookup": # find the associated user
         {
           "from": "users",
           "localField": "athlete_id",
           "foreignField": "user_id",
           "as": "user"
         }
    }
]


import pprint
joined_activity = list(db.activity.aggregate(join_pipeline))
pprint.pprint(joined_activity)
```
